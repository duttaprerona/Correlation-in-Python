### Importing Libraries

```python
# Importing Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

<a id="Data Cleaning"></a>
### Data Cleaning

```python
# Importing the data
df = pd.read_csv(r"../input/movies/movies.csv")
```

```python
# Reading the data
df.head(10)
```

```python
# Checking for Null Values
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{}-{}%'.format(col,pct_missing))
```

```python
# Dropping null values
df = df.dropna(how='any',axis=0) 
```

```python
# Checking data types of columns
df.dtypes
```

```python
# Changing formats of gross and budget
df['budget'] = df['budget'].astype('Int64')
df['gross'] = df['gross'].astype('Int64')
```

```python
df.head()
```

```python
# Creating correct year column
df['yearcorrect'] = df['released'].astype(str).str.split(', ').str[-1].astype(str).str[:4]
```

```python
df.head()
```

```python
# Order by Gross
pd.set_option('display.max_rows', 100)
df.sort_values(by= ['gross'], inplace= False, ascending= False)
```

```python
# Drop Any Duplicates
df['company'].drop_duplicates().sort_values(ascending=True)
```

<a id="Assumptions"></a>
### Assumptions
1. Budget has strong correlation with gross
2. Genre has strong correlation with gross
3. Company has strong correlation with gross

```python
# Scatter plot with Budget vs Gross Revenue
plt.scatter(x= df['budget'], y= df['gross'], color= 'orange')

plt.title('Budget vs Gross Earnings')

plt.xlabel('Budget')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
# Checking the correlation
df.corr()
```

```python
# Using a heatmap to show correlation of numeric variables
correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Numeric Features")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
# Changing object columns into numerical cat codes for further analysis
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name]= df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized
```

```python
# Finding correlation of new dataframe
df_numerized.corr()
```

```python
# Vizualising the new data frame using Heatmap
correlation_matrix = df_numerized.corr()

sns.set(rc = {'figure.figsize':(15,15)})
sns.heatmap(correlation_matrix, annot = True)

plt.title("Correlation matrix for Movies")

plt.xlabel("Movie features")

plt.ylabel("Movie features")

plt.show()
```

```python
#unstack and sort correlation

sorted_pairs = correlation_matrix.unstack().sort_values()

sorted_pairs
```

```python
#Finding high correlation between two variables
high_corr = sorted_pairs[(sorted_pairs) > 0.5]

high_corr
```
